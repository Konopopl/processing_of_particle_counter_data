```python
def number(data): 
    m = [0,31,29,31,30,31,30,31,31,30,31,30,31]  
    day=data.split(".")[0]  
    month=data.split(".")[1]  
    number=0 
    for i in range(int(month)): 
        number+=m[i] 
    number=number+int(day)

    #print(number) 
    return(number)
```
